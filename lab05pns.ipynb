```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_data(file_path):
 df=pd.read_csv("/content/employee_data_250.csv")
 return df
```

```python
def plot_histogram(df,column,bins=10):
 plt.figure(figsize=(8,6))
 sns.histplot(df[column],bins=bins,kde=True,color='blue',edgecolor='black')
 plt.title(f'Histogram of {column}')
 plt.xlabel(column)
 plt.ylabel('Frequency')
 plt.show()
```

```python
def plot_scatter(df,x_xolumn,y_column):
  plt.figure(figsize=(8,6))
  sns.scatterplot(x=df[x_column],y=df[y_column],color='orange')
  plt.title(f'Scatter Plot of{x_column} vs {y_column}')
  plt.xlabel(x_column)
  plt.ylabel(y_column)
  plt.show()
```

```python
def plot_bar_chart(df,x_column,y_column):
  plt.figure(figsize=(8,6))
  sns.barplot(x=df[x_column],y=df[y_column],color='green')
  plt.title(f'Bar Chart of {x_column} vs {y_column}')
  plt.xlabel(x_column)
  plt.ylabel(y_column)
  plt.xticks(rotation=45)
  plt.show()
```

```python
file_path='/content/employee_data_250.csv'
df=load_data(file_path)
```

```python
def plot_scatter(df, x_column, y_column): # Changed x_xolumn to x_column
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df[x_column], y=df[y_column], color='orange')  # Use x_column
    plt.title(f'Scatter Plot of {x_column} vs {y_column}')  # Use x_column
    plt.xlabel(x_column)  # Use x_column
    plt.ylabel(y_column)
    plt.show()
```

```python
plot_histogram(df,'Age')
plot_scatter(df,'Age','Salary')
plot_bar_chart(df,'Department','Salary')
```
